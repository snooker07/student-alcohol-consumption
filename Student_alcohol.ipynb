## Issue 5

### Define

Fix the erronous data types:
- SchoolEducationalSupport
- FamilyEducationalSupport      
- ExtraPaidClassesInCourse 
- ExtraCurricularActivities
- AttendedNurserySchool
- WantsHigherEducation
- HomeInternetAccess
- InRomanticRelationship

### Code

```python
# create string to boolean converter
def StringToBoolean(_str):
    return _str == "yes"
```

```python
# Fix boolean types
df_clean["SchoolEducationalSupport"] = df_clean["SchoolEducationalSupport"].apply(lambda x: bool(util.strtobool(x)))
df_clean["FamilyEducationalSupport"] = df_clean["FamilyEducationalSupport"].apply(lambda x: bool(util.strtobool(x)))
df_clean["ExtraPaidClassesInCourse"] = df_clean["ExtraPaidClassesInCourse"].apply(lambda x: bool(util.strtobool(x)))
df_clean["ExtraCurricularActivities"] = df_clean["ExtraCurricularActivities"].apply(lambda x: bool(util.strtobool(x)))
df_clean["AttendedNurserySchool"] = df_clean["AttendedNurserySchool"].apply(lambda x: bool(util.strtobool(x)))
df_clean["WantsHigherEducation"] = df_clean["WantsHigherEducation"].apply(lambda x: bool(util.strtobool(x)))
df_clean["HomeInternetAccess"] = df_clean["HomeInternetAccess"].apply(lambda x: bool(util.strtobool(x)))
df_clean["InRomanticRelationship"] = df_clean["InRomanticRelationship"].apply(lambda x: bool(util.strtobool(x)))
```
